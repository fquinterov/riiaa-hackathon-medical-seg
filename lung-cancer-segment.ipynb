```python
import numpy as np
from matplotlib import pyplot as plt

input = '/kaggle/input/lung-cancer-segment/train/20/data/130.npy'
output = '/kaggle/input/lung-cancer-segment/train/20/masks/130.npy'

input_img = np.load(input)
output_img = np.load(output)

plt.imshow(input_img, cmap = 'gray')
plt.show()
plt.imshow(output_img, cmap = 'viridis')
plt.show()
```

```python
color_mask = np.zeros((*output_img.shape, 3))
color_mask[output_img == 1] = [0, 1, 0]
plt.imshow(input_img, cmap='gray')  # Display the original image in grayscale
plt.imshow(color_mask, alpha=0.4)  # Overlay the color mask with 50% opacity
plt.show()
```
