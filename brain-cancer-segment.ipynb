```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

input = '/kaggle/input/brain-tumor-segmentation/images/1.png'
output = '/kaggle/input/brain-tumor-segmentation/masks/1.png'

input_img = cv2.imread(input, cv2.IMREAD_GRAYSCALE)
output_img = cv2.imread(output, cv2.IMREAD_GRAYSCALE)

plt.imshow(input_img, cmap = 'gray')
plt.show()
plt.imshow(output_img, cmap = 'viridis')
plt.show()
```

```python
color_mask = np.zeros((*output_img.shape, 3))
color_mask[output_img == 255] = [0, 1, 0]
plt.imshow(input_img, cmap='gray')  # Display the original image in grayscale
plt.imshow(color_mask, alpha=0.4)  # Overlay the color mask with 50% opacity
plt.show()
```
